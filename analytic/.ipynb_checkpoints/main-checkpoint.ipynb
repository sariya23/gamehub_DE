## Anti library

```python
anti_library  = pd.read_json(os.path.join(base_path, "anti_library_service", "anti_library.ndjson"), lines=True)
```

```python
anti_library.head()
```

```python
anti_library.dtypes
```

```python
anti_library.isna().sum()
```

```python
anti_library["reason"].unique()
```

```python
len(pd.merge(users, anti_library, on="user_id", how="inner")) == len(pd.merge(users, anti_library, on="user_id", how="left"))
```

```python
no_anti = pd.merge(users, anti_library, on="user_id", how="left")
```

```python
no_anti[no_anti["game_id"].isna()]
```
